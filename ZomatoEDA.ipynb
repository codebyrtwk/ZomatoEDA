```python
final_df = pd.merge(df , df2 , on ="Country Code" , how = "left")
```

```python
final_df.head()
```

```python
# Changing Country_x to Country
final_df.rename(columns={"Country_x": "Country"}, inplace=True)
```

```python
# Doing some observations
final_df.Country.value_counts()
```

```python

country = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
country_val
```

### OBSERVATION 1 : INDIA HAS MOST USERS AND TRANSACTION and Canada has Least Zomato Transactions  IN >US > UK

```python
#top 3 country that uses zomato
plt.pie(country_val[:3] , labels = country[:3] , autopct = '%1.2f%%')
```

```python
rating = final_df.groupby(["Aggregate rating", "Rating color" , "Rating text"]).size().reset_index().rename(columns = {0 : "count"})
```

```python
sns.barplot(x = "Aggregate rating" , y = "count" ,hue = "Rating color", data = rating , palette = ['white' , 'red', 'orange' , 'yellow' , 'green' , 'green'])
```
